```python
from pyspark import SparkContext
from pyspark.sql import SQLContext

sc = SparkContext.getOrCreate()
sqlContext = SQLContext(sc)
```

```python
!wget https://raw.githubusercontent.com/bensadeghi/pyspark-churn-prediction/master/data/churn-bigml-20.csv
```

```python
!ls
```

```python
df = sqlContext.read.load(
    'churn-bigml-20.csv',
    format='csv',
    header='true',
    inferSchema='true'
)
```

```python
df.show(5)
```

```python
from pyspark.sql.functions import avg

df.groupBy("Area code") \
  .agg(avg("Total day calls").alias("media_total_day_calls")) \
  .show()
```

```python
from pyspark.sql.functions import avg

df.groupBy("Area code") \
  .agg(avg("Total day minutes").alias("media_total_day_minutes")) \
  .show()
```

```python
df.filter(df["Total night calls"] < 55).show()
```
